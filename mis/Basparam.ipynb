```python
from itertools import permutations

with open("farsi.txt", "r") as f:
    lines = f.read()
    
    
charachters = ['ا', 'س' , 'پ' , 'ب' , 'ر' , 'م']


print("Valid ones:")
for shf_str in permutations(charachters):
    if "".join(shf_str) in lines:
        print("".join(shf_str))
        

print("")
print("Permutations:")        
for shf_str in permutations(charachters):
    print("".join(shf_str))
```
